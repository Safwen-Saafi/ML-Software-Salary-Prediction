```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey_results_public.csv")
```

```python
df.head()
#display the first 5 rows
```

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()


#df = df.dropna(): This line removes all rows in the DataFrame df that have any missing values (NaN). The dropna() method by default drops rows containing any NaN values.
#df.isnull().sum(): This line checks the DataFrame df for any missing values after the dropna() operation. The isnull() method returns a DataFrame of the same shape as df, with True for NaN values and False otherwise. The sum() method then sums up the True values (which are equivalent to 1), giving a count of missing values for each column.
```

```python
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
df.info()
```

```python
df['Country'].value_counts()
#Each country's count
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
#You'll notice that somehow people typed in Israel as their country, which is crazy
#We're gonna replace Israel with Palestine
```

```python
df['Country'] = df['Country'].replace('Israel', 'Palestine')
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
ax.set_ylim(0, 2000000) 
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
#The small green line entoured with a blue rectangle is the median value of the salary in each country
#All of the points are actually outliers
#We're gonna keep only the salaries in the most condensed areas
#None of the boxes surpasses the 250k limit
```

```python
df = df[df["Salary"] <= 250000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)

#We're gonna clean the years of experience converting them into floats
```

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)

#Doing the same for the Educational Level
```

```python
df["EdLevel"].unique()
```

```python
#Almost cleaned all the data
#However there is still an issue
#The educational level is astring and our model only understands numbers so we gotta convert these strings into unique numbers
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
#Bachelor's Degree => 0, ...
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
X = df.drop("Salary", axis=1)
y = df["Salary"]

#We split our data into features x
#And targets y
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X.values, y.values)
```

```python
#Normally we'd split our data into trainiing and testing set but for simplicity we're not
```

```python
y_pred = linear_reg.predict(X.values)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
error
```

```python
#As you can see the error is still high, 46k on each predicted value
#We try another model
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X.values, y.values)
```

```python
y_pred = dec_tree_reg.predict(X.values)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X.values, y.values)
```

```python
y_pred = random_forest_reg.predict(X.values)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X.values, y.values)
```

```python

regressor = gs.best_estimator_

regressor.fit(X.values, y.values)
y_pred = regressor.predict(X.values)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
